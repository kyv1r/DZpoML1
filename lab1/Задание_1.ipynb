```python
# фикс
from numpy import nan

def convert_to_float(df, column_name):
    # Convert the column to string, replace non-numeric values and NaN with empty strings, then convert to float
    df[column_name] = df[column_name].astype(str).str.replace('[^0-9.]', '', regex=True).replace('', nan).astype(float)
    return df[column_name]

# Example for replace the column dtype
df['Violent crime'] = convert_to_float(df, 'Violent crime')
df['Motor vehicle theft'] = convert_to_float(df, 'Motor vehicle theft')
df['Motor vehicle theft']
```

##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
#Агрегация
import statistics
from collections import defaultdict

# Создайте словарь для хранения данных по агентствам
dic = defaultdict(list)

# Предположим, что у вас уже есть DataFrame df
# Вы можете использовать df, который вы загрузили в предыдущем коде

# Агрегируйте данные по 'Agency' с вычислением медианы и среднего значения для 'Violent crime'
aggregated_df = df.groupby('Agency')['Violent crime'].agg(['median', 'mean'])

# Пройдите по агрегированным данным и добавьте их в словарь
for agency, row in aggregated_df.iterrows():
    median_value = row['median']
    mean_value = row['mean']
    dic[agency].append((median_value, mean_value))

# Выведите результаты
for agency, values in dic.items():
    for median_value, mean_value in values:
        print(f'{agency} - Медиана: {median_value}, Среднее: {mean_value}')
```
