##Поиск выбросов и аномалий

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
import numpy as np
import math
from sklearn.cluster import DBSCAN
from scipy import stats

class Outline:
    def __init__(self, data):
        # Инициализация класса с данными
        self.data = data

    # Метод обнаружения выбросов на основе среднего и стандартного отклонения
    def detect_outlier_mean_std(self, column_name, threshold=1.5):
        # Вычисление среднего и стандартного отклонения
        mean = self.data[column_name].mean()
        std = self.data[column_name].std()

        # Определение границ для выбросов
        lower_bound = mean - threshold * std
        upper_bound = mean + threshold * std

        # Определение выбросов
        outliers = (self.data[column_name] < lower_bound) | (self.data[column_name] > upper_bound)
        return self.data[outliers]

    # Метод обнаружения выбросов на основе межквартильного размаха
    def detect_outlier_kvart(self, column_name, threshold=1.5):
        # Вычисление квартилей
        Q1 = self.data[column_name].quantile(0.25)
        Q3 = self.data[column_name].quantile(0.75)

        # Вычисление межквартильного размаха
        IQR = Q3 - Q1

        # Определение границ для выбросов
        lower_threshold = Q1 - threshold * IQR
        upper_threshold = Q3 + threshold * IQR

        # Определение выбросов
        outliers = (self.data[column_name] < lower_threshold) | (self.data[column_name] > upper_threshold)
        return self.data[outliers]

    # Метод обнаружения выбросов на основе метода Шовене
    def detect_outlier_shovene(self, column_name):
        ans = []
        n = self.data[column_name].count()
        mean = self.data[column_name].mean()
        std = self.data[column_name].std()

        # Проверка точек на наличие выбросов с использованием метода Шовене
        for index, value in self.data.iterrows():
            if math.erfc((abs(value[column_name] - mean)) / std) < 1 / (2 * n):
                ans.append(index)
            else:
                n -= 1
        return ans

    # Метод обнаружения выбросов на основе алгоритма DBSCAN
    def detect_outlier_dbscan(self, col1, col2, eps, min_samples):
        # Преобразование данных в numpy массив
        column_data = self.data[[col1, col2]].to_numpy()

        # Применение алгоритма DBSCAN
        dbscan = DBSCAN(eps=eps, min_samples=min_samples).fit(column_data)
        labels = dbscan.labels_

        # Вычисление количества элементов в каждом кластере
        unique, counts = np.unique(labels, return_counts=True)
        clusters_counts = dict(zip(unique, counts))

        # Поиск кластеров с количеством элементов <= 2
        small_clusters = [cluster for cluster, count in clusters_counts.items() if count <= 2]

        # Возвращение индексов строк для этих кластеров
        outlier_indices = self.data.index[np.isin(labels, small_clusters)].tolist()

        return outlier_indices
```

Реализовать расчет корреляции Спирмана

```python
# Функция для расчета коэффициента корреляции Спирмена
from scipy.stats import spearmanr

def corr_spearman(column_1, column_2):
    # Вычисление коэффициента корреляции Спирмена и p-value
    rho, p = spearmanr(column_1, column_2)
    return rho
```

```python
# Пример использования функции и объяснение принципа работы
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}

# Создание DataFrame из данных
df = pd.DataFrame(data)

# Вызов функции corr_spearman для вычисления корреляции Спирмена между столбцами "Рост" и "Вес"
correlation = corr_spearman(df["Рост"], df["Вес"])

# Корреляция Спирмена измеряет степень монотонной связи между двумя переменными.
# Значение корреляции варьируется от -1 до 1.
# - Значение близкое к 1 указывает на положительную монотонную связь: при увеличении одной переменной увеличивается и другая.
# - Значение близкое к -1 указывает на отрицательную монотонную связь: при увеличении одной переменной уменьшается другая.
# - Значение близкое к 0 указывает на отсутствие монотонной связи.

# Этот метод быстро и удобно вычисляет корреляцию Спирмена между двумя столбцами данных,
# не требуя ручного вычисления рангов и использования специальных формул.
```

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```
